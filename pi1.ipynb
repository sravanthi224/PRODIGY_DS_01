# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Loading the Dataset

```python
df=pd.read_csv("population.csv")
```

```python
df.head()
```

# Bar Chart

```python
state_counts = df['State'].value_counts().sort_index()
print(state_counts)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(state_counts.index, state_counts.values, color='skyblue')
plt.title('Distribution of States')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Selecting the columns for the bar chart
County = df['County']
total_population = df['TotalPop']
```

```python
# Creating the bar chart
plt.figure(figsize=(10, 6))
plt.bar(County, total_population, color='skyblue')
plt.title('Total Population by County in Autauga County, Alabama')
plt.xlabel('County')
plt.ylabel('Total Population')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Histogram

```python
# Selecting the column for the histogram
mean_commute_times = df['MeanCommute']
```

```python
# Creating the histogram
plt.figure(figsize=(10, 6))
plt.hist(mean_commute_times, bins=5, color='skyblue', edgecolor='black')

# Adding title and labels
plt.title('Distribution of Mean Commute Times in Autauga County, Alabama')
plt.xlabel('Mean Commute Time (minutes)')
plt.ylabel('Frequency')

# Display the histogram
plt.tight_layout()
plt.show()
```

# Bar Chart For Gender

```python
# Extracting the total population of men and women for the bar chart
genders = ['Men', 'Women']
gender_counts = [df['Men'].sum(), df['Women'].sum()]
```

```python
# Creating the bar chart for gender distribution
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.bar(genders, gender_counts, color=['blue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Total Population')
```
